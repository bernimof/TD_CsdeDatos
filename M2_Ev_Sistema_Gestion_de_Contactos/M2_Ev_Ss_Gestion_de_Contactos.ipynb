# **Sistema de Gestión de Contactos**

### **Crear Clase: Contacto**

```python
# Creación de Clase - Contacto
class Contacto:
    def __init__(self, nombre, telefono, correo, direccion): # Constructor que asigna atributos del contacto
      # Atributos del contacto
        self.nombre = nombre
        self.telefono = telefono
        self.correo = correo
        self.direccion = direccion

    # Método para actualizar la información de un contacto
    def actualizar(self, nombre=None, telefono=None, correo=None, direccion=None):
        if nombre is not None:
            self.nombre = nombre
        if telefono is not None:
            self.telefono = telefono
        if correo is not None:
            self.correo = correo
        if direccion is not None:
            self.direccion = direccion

    # Método para mostrar contactos registrados
    def __str__(self):
        return f"Nombre: {self.nombre}, Teléfono: {self.telefono}, Correo: {self.correo}, Dirección: {self.direccion}"
```

### **Crear Clase: SistemaGestionContactos**

```python
# Creación de Clase - Sistema Gestion Contactos
class SistemaGestionContactos:
    def __init__(self): # Constructor
        self.contactos = [] # Comienza con una lista vacía.

#  Método para agregar un nuevo contacto al sistema.
    def agregar_contacto(self, contacto):
        self.contactos.append(contacto)

#  Método para eliminar un contacto por nombre.
    def eliminar_contacto(self, nombre):
      self.contactos = [contacto for contacto in self.contactos if contacto.nombre != nombre]

#  Método para editar un contacto existente por su nombre.
    def editar_contacto(self, nombre_a_editar, nuevo_nombre=None, nuevo_telefono=None, nuevo_correo=None, nueva_direccion=None):
        for contacto in self.contactos:
            if contacto.nombre == nombre_a_editar:
                contacto.actualizar(nuevo_nombre, nuevo_telefono, nuevo_correo, nueva_direccion)
                print(f"Contacto con nombre {nombre_a_editar} editado exitosamente.")
                return
        print(f"Contacto con nombre {nombre_a_editar} no encontrado.")

#  Método para buscar un contacto por nombre o teléfono.
    def buscar_contacto(self, parametro):
        return [contacto for contacto in self.contactos if parametro.lower() in contacto.nombre.lower() or parametro in contacto.telefono]

#  Método para mostrar todos los contactos en el sistema.
    def mostrar_contactos(self):
        if not self.contactos:
            print("No hay contactos en el sistema.")
        else:
            for contacto in self.contactos:
                print(contacto)
```

### **Creación de Aplicación**

```python
# Creación de Aplicación
def menu():
    gestor = SistemaGestionContactos()

    # Menú del sistema
    while True:
        print("\n Menú del Sistema de Gestión de Contactos")
        print("1. Agregar contacto")
        print("2. Editar contacto")
        print("3. Eliminar contacto")
        print("4. Buscar contacto")
        print("5. Mostrar todos los contactos")
        print("6. Salir")
        opcion = input("Selecciona una opción: ")
        print("\n")

        # Agregar contacto
        if opcion == '1':
            print("Ingresé la información del contacto nuevo a agregar:")
            nombre = input("Nombre: ")
            telefono = input("Teléfono: ")
            correo = input("Correo: ")
            direccion = input("Dirección: ")
            contacto = Contacto(nombre, telefono, correo, direccion)
            gestor.agregar_contacto(contacto)
            print("Contacto agregado.\n")

        # Editar contacto
        elif opcion == '2':
            nombre_a_editar = input("Ingresa el nombre del contacto a editar: ")

            # Validación de que el contacto existe
            if not any(contacto.nombre == nombre_a_editar for contacto in gestor.contactos):
                print("Contacto no encontrado. Intente nuevamente.\n")
                continue

            # Pedir nuevos datos
            nuevo_nombre = input("Nuevo nombre (dejar vacío para no cambiar): ")
            nuevo_telefono = input("Nuevo teléfono (dejar vacío para no cambiar): ")
            nuevo_correo = input("Nuevo correo (dejar vacío para no cambiar): ")
            nueva_direccion = input("Nueva dirección (dejar vacío para no cambiar):")

            # Actualiza el contacto
            gestor.editar_contacto(nombre_a_editar, nuevo_nombre or None, nuevo_telefono or None, nuevo_correo or None, nueva_direccion or None)
            print("Contacto actualizado. \n")

        # Eliminar contacto
        elif opcion == '3':
            nombre = input("Nombre del contacto a eliminar: ")

            # Verificar si el contacto existe
            if not any(contacto.nombre == nombre for contacto in gestor.contactos):
                print("Contacto no encontrado.\n")
                continue

            # Elimina el contacto
            gestor.eliminar_contacto(nombre)
            print("Contacto eliminado.\n")

        # Buscar contacto
        elif opcion == '4':
            parametro = input("Buscar por nombre o teléfono: ")

            # Buscar coincidencia
            resultados = gestor.buscar_contacto(parametro)
            if resultados:
                for contacto in resultados:
                    print(f"{contacto.nombre}, {contacto.telefono}, {contacto.correo}, {contacto.direccion}")
            else:
                print("No se encontró ningún contacto que coincida.\n")

        # Mostrar todos los contactos
        elif opcion == '5':
            print("\n Contactos en el sistema:")
            gestor.mostrar_contactos()
            print("\n")

        # Salir del menú
        elif opcion == '6':
            print("Saliendo del Sistema de Gestión de Contactos. ¡Hasta luego! :)")
            break
        else:
            print("Opción no válida. Ingresé una opción del Menú. \n")
```

### **Pruebas Unitarias == Prototipo funcional**

```python
# Activa la Aplicación
if __name__ == "__main__":
    menu()
```
